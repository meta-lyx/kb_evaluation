```python
import pandas as pd
import math
import os

def split_dataset(file_path, num_splits=5):
    # 创建输出目录
    output_dir = 'split_datasets'
    os.makedirs(output_dir, exist_ok=True)
    
    # 计算每个分片的大小
    total_lines = sum(1 for _ in open(file_path))
    chunk_size = math.ceil(total_lines / num_splits)
    
    print(f"Total lines: {total_lines}")
    print(f"Chunk size: {chunk_size}")
    
    # 使用 chunksize 参数分批读取数据
    for i, chunk in enumerate(pd.read_csv(file_path, 
                                        sep='\t',
                                        header=None,
                                        names=['doc_id', 'text'],
                                        chunksize=chunk_size)):
        output_path = os.path.join(output_dir, f'fulldocs_part_{i+1}.tsv')
        chunk.to_csv(output_path, sep='\t', index=False)
        print(f"Saved part {i+1} with {len(chunk)} rows to {output_path}")

# 执行分割
split_dataset('datasets/fulldocs.tsv')
```

```python
import gzip
import shutil
import os
import pandas as pd
from beir import util


# 选择 MS MARCO 数据集
dataset = "msmarco"
url = f"https://msmarco.z22.web.core.windows.net/msmarcoranking/fulldocs.tsv.gz"

# 下载数据集
save_dir = "datasets"
os.makedirs(save_dir, exist_ok=True)
data_path = os.path.join(save_dir, "fulldocs.tsv.gz")
util.download_url(url, data_path)

# 解压 .gz 文件
extracted_file_path = data_path.replace('.gz', '')
with gzip.open(data_path, 'rb') as f_in:
    with open(extracted_file_path, 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)
```

```python
# Load qrel.train.tsv (query relevance file)
qrels_path = "/Users/yingxuanli/Downloads/ncs_hk/ncs_contract_review/qrels.train.tsv"

qrels = {}
with open(qrels_path, "r", encoding="utf-8") as file:
    for line in file:
        query_id, _, doc_id, relevance = line.strip().split("\t")
        if query_id not in qrels:
            qrels[query_id] = []
        if int(relevance) > 0:  # Only consider relevant documents
            qrels[query_id].append(doc_id)

print(f"Loaded {len(qrels)} query-document relevance mappings.")
```
